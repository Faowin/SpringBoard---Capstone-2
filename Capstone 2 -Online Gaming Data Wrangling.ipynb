```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('online_gaming_behavior_dataset.csv')
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
#change InGamePurchases 0 = No 1 = Yes
df['InGamePurchases'] = np.where(df['InGamePurchases'] == 0, 'No', 'Yes')
df.head()
```

```python
#groupby GameGenre
byGameGenreMeans = df.groupby('GameGenre')[['Age','PlayTimeHours','SessionsPerWeek','AvgSessionDurationMinutes', 'PlayerLevel', 'AchievementsUnlocked']].mean()
byGameGenreCounts = df.groupby('GameGenre')['PlayerID'].count()
print(byGameGenreMeans.head())
print(byGameGenreCounts.head())
```

```python
#groupby GameDifficulty
byGameDifficultyMeans = df.groupby('GameDifficulty')[['Age','PlayTimeHours','SessionsPerWeek','AvgSessionDurationMinutes', 'PlayerLevel', 'AchievementsUnlocked']].mean()
byGameDifficultyCounts = df.groupby('GameDifficulty')['PlayerID'].count()
print(byGameDifficultyMeans.head())
print(byGameDifficultyCounts.head())
```
